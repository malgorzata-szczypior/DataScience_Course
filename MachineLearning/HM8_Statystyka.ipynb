**Wykonaj wykres rozkładu prawdopodobieństwa i dystrybuanty tej zmiennej losowej.**

```python
wynik=set(range(1,7))
rzuty=set(it.product(wynik,repeat=2))
all=len(rzuty) #moc zbioru wartości elementarnych
zmienna =[]
for i in rzuty:
    if i[0]==i[1]:
        zmienna.append(0)
    if i[0] <i[1]:
        zmienna.append(1)
    if i[0]> i[1]:
        zmienna.append(2)
```

```python
suma = dict()
for el in zmienna:
    suma[el]=zmienna.count(el)
suma
```

**3. Dana jest zmienna losowa o rozkładzie normalnym ze średnią $\mu=1250$ i odchyleniem standardowym $\sigma=300$. Ta zmienna losowa opisuje sumę dziennego przychodu pewnej firmy w PLN. Oblicz:**
 
 1. Prawdopodobieństwo, że w losowo wybranym dniu suma dziennego przychodu osiągnie wartość pomiędzy $1000$ a $1100$ PLN
 2. Prawdopodobieństwo, że w losowo wybranym dniu suma dziennego przychodu osiągnie wartość większą niż $1500$ PLN
 3. Kwotę przychodu dziennego, której prawdopodobieństwo przewyższenia (otrzymania wartości równej lub większej) wynosi $0,1$
 4. Kwotę przychodu dziennego, której prawdopodobieństwo nieosiągnięcia (otrzymania wartości równiej lub niższej) wynosi $0.25$
 
**Wykonaj wykres gęstości rozkładu prawdopodobieństwa i dystrybuanty tej zmiennej losowej. Wykonaj wykresy ilustrujące rozwiązanie dla każdego przypadku**

```python
m=1250
s=300
rozkladnorm=st.norm(m,s)
```

```python
#rozkład prawdopodobieństwa
r=5
x=np.arange(m-r*s,m+r*s,0.25)
f=rozkladnorm.pdf(x)
plt.plot(x,f)
plt.legend(["f(x) - gęstość"])
```

```python
#dystrybuanta
F=rozkladnorm.cdf(x)
plt.plot(x,F)
plt.legend(["F(x) - dystrybuanta"])
```

```python
#1 1000-1100
a=1000
b=1100
p1=rozkladnorm.cdf(b) - rozkladnorm.cdf(a)
p1
plt.plot(x,f,"blue")
t = np.linspace(a,b)
plt.fill_between(t,0,rozkladnorm.pdf(t))
plt.legend(["f(x) - gęstość", "P(a<X<b)"])
pass
```

```python
#2 1500
c=1500
p2=1-rozkladnorm.cdf(c)
p2
plt.plot(x,f,"blue")
t = np.linspace(c,2750,endpoint=False)
plt.fill_between(t,0,rozkladnorm.pdf(t))
plt.legend(["f(x) - gęstość", "P(X>c)"])
pass
```

```python
#3 0.1
p = 0.1
q = rozkladnorm.ppf(p)
print(q)

plt.plot(x,F,"blue")

t = np.linspace(q,m+5*s,100)
plt.fill_between(t,0,rozkladnorm.cdf(t))
plt.legend(["f(x) - gęstość", "P(X>q)=0,1"])
pass
```
